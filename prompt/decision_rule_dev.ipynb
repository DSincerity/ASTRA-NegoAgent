```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid_scale(x, alpha=1.0, mu=0.5):
    return 1 / (1 + np.exp(-alpha * (x - mu)))

def quadratic_scale(x, mu=0.5):
    return (x - mu)**2

# Generate a range of original values
original_values = np.linspace(0, 1, 100)

# Apply the scaling functions
sigmoid_scaled_values = sigmoid_scale(original_values, alpha=10, mu=0.5)
quadratic_scaled_values = quadratic_scale(original_values, mu=0.5)

# Plot the results
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(original_values, sigmoid_scaled_values, label='Sigmoid Scale (alpha=10, mu=0.5)')
plt.title('Sigmoid Scaling')
plt.xlabel('Original Value')
plt.ylabel('Scaled Value')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(original_values, quadratic_scaled_values, label='Quadratic Scale (mu=0.5)')
plt.title('Quadratic Scaling')
plt.xlabel('Original Value')
plt.ylabel('Scaled Value')
plt.legend()

plt.tight_layout()
plt.show()
```
